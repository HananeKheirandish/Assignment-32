```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 100
std = 4

melon_width = np.random.normal(8, std, N)
melon_length = np.random.normal(20, std, N)
melon_weight = np.random.normal(2, 0.8, N)
# melons = np.zeros(N, dtype='int')

balloon_width = np.random.normal(40, std, N)
balloon_length = np.random.normal(40, std, N)
balloon_weight = np.random.normal(0.01, 0.008, N)
# balloons = np.ones(N, dtype='int')
```

```python
width = np.concatenate((melon_width, balloon_width))
length = np.concatenate((melon_length, balloon_length))
weight = np.concatenate((melon_weight, balloon_weight))
# name = np.concatenate((melons, balloons))

data = np.array([width, length, weight]).T
```

```python
plt.scatter(data[0:N, 0], data[0:N, 1], data[0:N, 2],
            c = 'green', label = 'melons')

plt.scatter(data[N: , 0], data[N: , 1], 100*data[N: , 2], 
            c = 'purple', label = 'balloons')

plt.xlabel('Width')
plt.ylabel('Length')
plt.legend()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')

ax.scatter(data[0:N, 0], data[0:N, 1], data[0:N, 2],
            c = 'green', label = 'melons')

ax.scatter(data[N: , 0], data[N: , 1], data[N: , 2], 
            c = 'purple', label = 'balloons', marker='^')

ax.set_xlabel("width")
ax.set_ylabel("length")
ax.set_zlabel("weight")
```
